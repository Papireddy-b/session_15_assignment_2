# Importing required libraries

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
import re
from collections import Counter
from nltk.corpus import stopwords
```

```python
import re
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plt
import seaborn as sns
```

# Extracting text from Webpage

```python
response =urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

# Extracting words from text

```python
text = soup.get_text()
```

```python
print(text)
```

```python
# Create tokenizer
tokenizer = RegexpTokenizer('\w+')

# Create tokens
tokens = tokenizer.tokenize(text)
tokens[:5]
```

```python
nltk.download('stopwords')
```

```python
sw = nltk.corpus.stopwords.words('english')
print(sw)
```

```python
# Lowercase all words 
words = [word.lower() for word in tokens]
```

```python
words[:15]
```

# Removing stopwords


```python
words = [word for word in words if word not in sw]
```

```python
print(len(words))
```

# Removing numbers


```python
words = [word for word in words if not word.isnumeric()]
```

```python
print(len(words))
```

# Remove single-character tokens (mostly punctuation)


```python
words = [word for word in words if len(word) > 1]
```

```python
print(len(words))
```

```python
# Figures inline and set visualization style
%matplotlib inline
sns.set()

# Create freq dist and plot
freqdist1 = nltk.FreqDist(words)
freqdist1.plot(25)
```

# Frequency of words in Webpage

```python
#freqdist1.most_common()
for word, count in freqdist1.most_common():
    print(u'{}:{}'.format(word, count))
```

We can see the frequency of words in the webpage above
